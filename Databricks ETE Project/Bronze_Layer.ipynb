### Parameters

```python
dbutils.widgets.text('file_name', '')
```

```python
param_file_name = dbutils.widgets.get('file_name')
```

```python
print(param_file_name)
```

### Data Reading


```python
df = (
    spark
    .readStream
    .format('cloudFiles')
    .option('cloudFiles.format', 'parquet')
    .option('cloudFiles.schemaLocation', f'abfss://bronze@databricksstorageete.dfs.core.windows.net/checkpoint_{param_file_name}')
    .load(f'abfss://source@databricksstorageete.dfs.core.windows.net/{param_file_name}')
)
```

### Data Writing

```python
(df
 .writeStream
 .format('parquet')
 .outputMode('append')
 .option('checkpointLocation', f'abfss://bronze@databricksstorageete.dfs.core.windows.net/checkpoint_{param_file_name}')
 .option('path', f'abfss://bronze@databricksstorageete.dfs.core.windows.net/{param_file_name}')
 .trigger(once=True)
 .start()
)
```
