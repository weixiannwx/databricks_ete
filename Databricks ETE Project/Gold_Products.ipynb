### DLT Pipeline

```python
import dlt
import pyspark.sql.functions as F
import pyspark.sql.types as T
```

#### Streaming Table

```python
# Expectations
rules_to_test = {
    'rule1': 'product_id IS NOT NULL',
    'rule2': 'product_name IS NOT NULL'
}
```

```python
@dlt.table()
@dlt.expect_all_or_drop(rules_to_test)
def DimProducts_stage():
  return (
    spark.readStream.table('databricks_cata.silver.products')
  )
```

```python
@dlt.view()
def DimProducts_view():
  return (
    spark.readStream.table('Live.DimProducts_stage')
  )
```

#### DimProductts

```python
dlt.create_streaming_table('DimProducts')
```

```python
dlt.apply_changes(
    target = 'DimProducts',
    source = 'Live.DimProducts_view',
    keys = ['product_id'],
    sequence_by='update_date',
    stored_as_scd_type='2'
)
```
